# KMeans on Iris Dataset

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = load_iris()
```

```python
data
```

```python
data.target
```

```python
data.data
```

```python
data_df = pd.DataFrame(data.data, data.target)
```

```python
data_df
```

```python
data_df.rename(columns= {0 : 'sepal length', 1 : 'sepal width', 2 : 'petal length', 3 : 'petal width' }, inplace= True)
```

```python
data_df
```

```python
data_df.drop(columns= ['petal length', 'petal width'], inplace= True)
```

```python
data_df
```

```python
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.scatter(data_df['sepal length'], data_df['sepal width'])
```

```python
km = KMeans(n_clusters= 5)
```

```python
predicted = km.fit_predict(data_df[['sepal length', 'sepal width']])
```

```python
data_df['cluster'] = predicted
```

```python
data_df
```

```python
km.cluster_centers_
```

```python
data_clust0 = data_df[data_df.cluster== 0]
data_clust1 = data_df[data_df.cluster== 1]
data_clust2 = data_df[data_df.cluster== 2]
data_clust3 = data_df[data_df.cluster== 3]
data_clust4 = data_df[data_df.cluster== 4]
```

```python
data_clust0
```

```python
plt.scatter(data_clust0['sepal length'], data_clust0['sepal width'], marker= '+', color= 'green')
plt.scatter(data_clust1['sepal length'], data_clust1['sepal width'], marker= ',', color= 'red')
plt.scatter(data_clust2['sepal length'], data_clust2['sepal width'], marker= '.', color= 'blue')
plt.scatter(data_clust3['sepal length'], data_clust3['sepal width'], marker= '.', color= 'gold')
plt.scatter(data_clust4['sepal length'], data_clust4['sepal width'], marker= ',', color= 'yellow')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color= 'black', marker= '*', label= 'Centroid')
plt.legend()
```

## Elbow Testing

```python
sse = []
k_rng = range(1, 10)
for k in k_rng:
    km = KMeans(n_clusters= k)
    km.fit(data_df[['sepal length', 'sepal width']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
