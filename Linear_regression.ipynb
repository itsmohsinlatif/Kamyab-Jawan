```python
import pandas as pd
```

```python
income_data = pd.read_csv('canada_per_capita_income.csv')
```

```python
income_data
```

```python
from sklearn import linear_model
```

```python
year = income_data.drop(['per capita income (US$)'], axis= 1)
```

```python
year
```

```python
income = income_data.drop('year', axis= 1)
```

```python
income
```

```python
lin_reg = linear_model.LinearRegression()
```

```python
lin_reg.fit(year, income)
```

```python
predicted_income = lin_reg.predict(year)
```

```python
predicted_income
```

```python
#Cofficent
print('Coefficent: ', lin_reg.coef_)
```

```python
#Mean_square_error
from sklearn.metrics import mean_squared_error
print("Mean Square Error: ", mean_squared_error(income, predicted_income))
```
