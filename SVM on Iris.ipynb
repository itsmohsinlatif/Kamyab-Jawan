```python
#adding TARGET COLUMN TO THE DATAFRAME

iris_df['target_values'] = iris_data.target
```

## Training Phase

#Data splitting into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x = iris_df.drop(['target_values', 'target_names'], axis= 'columns')
y = iris_data.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2)
```

```python
#checking the instences in train and test

len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
```

```python
#Model for training
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict([[4.8, 3.0, 1.5, 0.3]])
```

```python
model.predict([[0.8, 3.0, 1.0, 2.3]])
```
